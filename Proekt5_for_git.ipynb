Оказалось, что помимо пропущенных строк в этом столбце имеется такое значение как "tbd".

В контексте оценок компьютерных игр, аббревиатура "TBD" обычно означает "To Be Determined", что переводится как "Будет определено позже".Такая формулировка имеет место быть, когда игра ещё не вышла или не собрала достаточное количество отзывов, чтобы можно было выставить оценку.

Так как анализируемые данные за предшествующие годы, отсутсвует вероятность, что та или иная игра еще не вышла и не успела собрать оценки. Соотвтетственно такие игры в принципе не набрали достаточное количество отзывов, а значит эти значения необходимо обрабатывать как пропущенные данные.

Проверим сколько строк содержит значение "tbd".

```python
x=data.query('user_score == "tbd"')['user_score'].count()
y=data['user_score'].isnull().sum()
z=data['platform'].count()

display(Markdown(
f"Значение ***tbd*** содержат {x} строк или {round(x/data['user_score'].count()*100,0).astype('int')}%.\
 Пропуски содержат {y} строк или {round(y/z*100,0).astype('int')}%.\n\n \
Пропусков обнаружено очень много- {round(y/z*100,0).astype('int')}% \
а если еще и учесть количество строк с ***tbd*** ({round(x/z*100,0).astype('int')}%), то в общей сложности это \
целых {round(x/z*100,0).astype('int')+round(y/z*100,0).astype('int')}%! \n\n \
Заменим пропуски на значение -1 и в дальнейшем, если потребуется анализ именно оценок пользователей, их можно \
будет быстро отфильтровать (так как прочие оценки от 0 и до 10)."))
data['user_score'] = data['user_score'].replace('tbd', np.NaN).astype('float').fillna(-1)
```

```python
# check
# пропущенные значения бары

def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            #Добавила сетку
            .plot(kind = 'barh', figsize = (15,6), rot = 0, legend = False, fontsize = 16, grid=True)
            #Изменила название
            .set_title('Наличие пропусков в столбцах' + "\n", fontsize = 20, color = 'SteelBlue')
        )
   
    except:
        print('пропусков не осталось :) или произошла ошибка в первой части функции ')
```

```python
def hist1(data,column,name,x,bins,range1):
    df = data.loc[data[column] != -1]
    df[column].plot(kind='hist',figsize=(10, 5), bins=bins, grid=True, range = range1)
    # Добавляем заголовок
    plt.title('Расспределение:' + name)
    # Добавляем подпись оси X
    plt.xlabel(x)
    plt.show()

columns_to_plot = [
    (data,'year_of_release', ' год выпуска', 'год',100, (1980, 2016)),
    (data,'na_sales', ' продажи в Северной Америке', 'кол-во, млн. копий',50,(0, 3.5)),
    (data,'eu_sales', ' продажи в Европе', 'кол-во, млн. копий',50,(0, 2.5)),
    (data,'jp_sales', ' продажи в Японии', 'кол-во, млн. копий',50,(0, 1.5)),
    (data,'other_sales', ' продажи в других регионах', 'кол-во, млн. копий',50,(0, 1.5)),
    (data,'critic_score', ' оценка критиков', 'оценка',30,(0, 100)),
    (data,'user_score', ' оценка пользователей', 'оценка',30,(0, 10))
]

for data, column, name, x, bins, range1 in columns_to_plot:
    hist1(data, column, name, x, bins, range1)
```

## Изменение продаж по платформам.

[(к содержанию)](#Содержание)

Перейдем к анализу продаж по платформам.

```python
data.groupby('platform')['all_sales'].sum().sort_values() \
                                     .plot(kind='bar', figsize = (15,6), grid=True, label='сумма продаж')
# Добавляем заголовок
plt.title('Диаграмма количества игр, выпущенных в разные годы на разных пллатформах')
# Добавляем подпись оси X и Y
plt.xlabel('виды платформ')
plt.ylabel('кол-во, млн. копий')
# Поворачиваем метки на оси X на 50 градусов для вертикального отображения
plt.xticks(rotation=50)
# Добавляем легенду
plt.legend()
plt.show()

display(Markdown(
f"График показывает, что {sum(data.groupby('platform')['all_sales'].sum()>400)} платформ сильно \
выделяются по уровню продаж среди прочих. При этом самой продаваемой \
можно назвать *PS2*."))

display(Markdown(
f"Посмотрим, как продавались самые популярные платформы по годам."))

# Отфильтровываем строки, где год выпуска не равен -1
filtered_data = data.query('year_of_release != -1')

# Группируем отфильтрованные данные по платформам и суммируем продажи
platform_sales = data.query('year_of_release != -1').groupby('platform')['all_sales'].sum()

# Фильтруем данные по платформам с продажами более 400
best_platforms_data = filtered_data[filtered_data['platform'].isin(platform_sales[platform_sales > 400].index)]

# Создаем сводную таблицу
best_platform_pivot = best_platforms_data.pivot_table(index='year_of_release',\
                                                columns='platform', \
                                                values='all_sales',\
                                                aggfunc='sum')
# Строим линейный график с маркерами
ax = best_platform_pivot.plot(kind='line', marker='o', figsize=(15,5))
# Устанавливаем форматировщик для оси X, чтобы отображались только целые числа
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
# Добавляем название графика и подписи осей
plt.title('Продажи популярных платформ по годам')
plt.xlabel('года')
plt.ylabel('кол-во, млн. копий')

plt.grid(True)

# Отображаем график
plt.show()
```

Из графика видно, что большинство платформ проходят через этапы роста, пика и спада в продажах. Средний срок жизни платформы можно оценить, рассмотрев промежуток времени от момента ее появления на графике до момента заметного спада продаж.

В среднем, основываясь на представленных данных, платформы кажутся актуальными в течение примерно 10 лет. При этом максимума популярности они набирают приблизительно в середине своего "жизненного цикла".

## Диаграммы размаха по продажам каждой из платформ.

[(к содержанию)](#Содержание)

```python
def box(data_box, col, type1, name, x, xlimit):
    dt = data_box.loc[data_box[col] != -1]
    grouped = dt.groupby(type1)
    if xlimit == 1:
        # Группируем данные по платформам и строим boxplot
        dt.boxplot(column=col, by=type1, vert=False, figsize=(15, 12), showfliers=False, whis=0)
        q3 = grouped[col].quantile(0.75).max()
        plt.xlim(0, q3+0.1)
    else: 
        dt.boxplot(column=col, by=type1, vert=False, figsize=(15, 12))
    # Добавляем заголовок
    plt.title('Диаграмма размаха: ' + name)
    # Убираем автоматический заголовок, созданный параметром by
    plt.suptitle('')
    # Добавляем подпись оси X
    plt.xlabel(x)
    plt.show()
    return 

box(data_new, 'all_sales','platform', 'продажи по платформам', 'количество, млн. копий',0)
```

# Проверка гипотез

[(к содержанию)](#Содержание)

## Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

[(к содержанию)](#Содержание)

Проверим следующие гипотезы:
- Нулевая гипотеза (Н0): Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
- Альтернативная гипотеза (Н1): Средние пользовательские рейтинги платформ Xbox One и PC разные

При формулировке гипотез придерживаемся правила, что нулевая гипотеза должна формулироваться так, чтобы проверялось равенство. Поэтому в ней в нашем случае мы говорим о равенстве средних пользовательских рейтингов платформ Xbox One и PC. Соответственно альтернативная говорит об обратном предположении - неравенстве.

Определим уровень статистической значимости равным 0.05. Это традиционный и наиболее распространенный уровень значимости. Если p-значение меньше 0.05, считается, что результаты эксперимента достаточно редки или необычны, чтобы отвергнуть нулевую гипотезу.

```python
#Создадим Series с данными о времени поездки 

dt_xbox_one = data_new[(data_new['platform'] == 'xone') & (data_new['user_score'] > 0)]['user_score']
dt_pc = data_new[(data_new['platform'] == 'pc') & (data_new['user_score'] > 0)]['user_score']

alpha = 0.05 # уровень статистической значимости 

# Проверяется двусторонняя гипотеза
results = st.ttest_ind(dt_xbox_one, dt_pc)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу о равенстве средних пользовательских рейтингов')
else:
    print('Не получилось отвергнуть нулевую гипотезу о равенстве средних пользовательских рейтингов')
```

P-значение измеряет вероятность наблюдения различий в данных (или более экстремальных различий), как минимум таких же, как те, что были фактически получены в эксперименте, если бы нулевая гипотеза была верной. Если эти экстремальные различия встречаются редко (p-значение низкое), это подразумевает, что наблюдаемый эффект может быть неслучайным и, следовательно, нулевая гипотеза может быть неверной.

В нашем случае, если бы нулевая гипотеза была верной (то есть разницы между средними пользовательскими оценками Xbox One и PC нет), p-значение 0.1045 означает, что вероятность получить различия в средних пользовательских оценках, которые наблюдались в исследовании, или более крупные различия, составляет примерно 10.45%, если бы на самом деле не было никакой разницы. 

На основе теста можем сделать вывод, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

## Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

[(к содержанию)](#Содержание)

Проверим следующие гипотезы:
- Нулевая гипотеза (Н0): Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые
- Альтернативная гипотеза (Н1): Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

При формулировке гипотез придерживаемся правила, что нулевая гипотеза должна формулироваться так, чтобы проверялось равенство. Поэтому в ней в нашем случае мы говорим о равенстве средних пользовательских рейтингов жанров Action и  Sports. Соответственно альтернативная гипотеза говорит об обратном предположении - неравенстве.

Определим уровень статистической значимости равным 0.05.

Необходимо произвести проверку о равенстве средних двух гинеральных совокупностей, поэтому будем использовать метод st.ttest_ind(). 

```python
#Создадим Series с данными о времени поездки 

dt_action = data_new[(data_new['genre'] == 'Action') & (data_new['user_score'] > 0)]['user_score']
dt_sports = data_new[(data_new['genre'] == 'Sports') & (data_new['user_score'] > 0)]['user_score']

alpha = 0.05 # уровень статистической значимости 

# Проверяется двусторонняя гипотеза
results = st.ttest_ind(dt_action, dt_sports)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу о равенстве средних пользовательских рейтингов')
else:
    print('Не получилось отвергнуть нулевую гипотезу о равенстве средних пользовательских рейтингов')
```

Тест показал, что средние пользовательские рейтинги жанров Action и Sports разные.

***Потенциально прибыльными платформами для 2017 года исходя из анализа будут PS4 и Xbox One, так как они показывают рост продаж и являются самыми современными платформами на данный момент.*** Помимо этих, отметим еще и ***платформу 3DS,*** объем продаж которой составляет около 20 млн. копий и ***платформу PS,*** уровень ее продаж пусть и не самый высокий, но на графике продажи по ней выглядят весьма стабильно.

Диаграммы размаха показывают, что платформы, такие как PS3 и X360, имеют большее количество игр с высокими продажами, что может отражать более крупную или более активную пользовательскую базу в прошлом. В то же время более новые системы, такие как PS4 и XOne, имеют меньший размах продаж, возможно потому как данных для них меньше.

Было определено, что наблюдается прямая зависимость количества продаж консолей от оценок критиков. При этом, можно сказать, что продажи по разным платформам имеют совершенно разную зависимость от оценки пользователей. Здесь наблюдается как сильная прямая зависимость(как у Wii или WiiU), так и хоть и очень слабая, но обратная зависимость(X360, PS4, XOne, PC).

В пятерку лидеров по доли каждого из жанров входят: Action, Role-Playing, Adventure, Sports, Shooter. Из них, жанр Action составляет 34%.
Лидирующие жанры по сумме продаж: Action, Shooter, Sports, Role-Playing, Misc. Сделан вывод, что спрос на игры в исследуемых жанрах не всегда соответствует объему выпуска этих игр. Так, жанр Misc не вошел в пятерку лидеров пообъему выпуска, но смог выбиться в лидеры по объемам продаж.
Однако после анализа диаграмм разброса самым перспективным и продаваемым жанром оказался ***Shooter*** с медианным значением в 0.515 миллионов копий. Так же можно отметить ***Sports и Platform,*** которые так же показывают хорошие значения средней прибыли(0.180, 0.140 млн. копий соответственно)

При определении портрета пользователя каждого из регионов были выявлены самые популярные платформы для различных регионов:

Сделан вывод, что в разных регионах популярны разные платформы.

Анализ жанров по регионам показал, что как и в случае с платформами, Японские пользователи имеют отличные от других регионов предпочтения. Их самым любимым жанром является Role-Playing, но жанр Action от него не сильно отстает.  В Америке и Европе это Shooter и Action.

При анализе влияния рейтинга игр на продажи было выявлено,  что в Японии основными потребителями игрового контента являются подростки от 13 лет, а в Америке и Европе - люди, старше 17 лет.

При этом, важно подчеркнуть, что общий объем продаж в Японии находится на самом низком уровне - 93.09 млн копий, в то время как у Америки и Европы это 284.06 и 270.68 соответственно.

- ***Типичный пользователь в Японии выбирает платформу 3DS, жанры игры Role-Playing и Action, имеет возраст от 13 лет.***
